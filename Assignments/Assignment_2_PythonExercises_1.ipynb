##1.Text Analysis Tool:
#### Create a Python program that reads a text file and performs various analyses on its content. The program should: Count the number of words, sentences, and paragraphs. Calculate the average word length. Identify the most common words and their frequencies.

```python
##Text Analysis Tool AMSG_1948005
import re
with open('archive.txt',  'r') as file:
  text = file.read()

  def word_counter(text): #funtion that counts the number of words
    words = text.split()
    return len(words)

  def sentence_counter (text): #funtion that counts sentences by using (., ! y ?)
    sentences = re.split(r'[.!?]', text)
    return len (sentences)

  def paragraph_counter (text): #funtion that counts the number of paragraphs
    paragraphs = text.split ('\n\n')
    return len (paragraphs)

  def average_word_lenght  (text): #function that calculates the average of the words lenght
   words = text.split()
   words_lenght = [len(word) for word in words]
   average_word_lenght =sum(words_lenght)/(word_counter(text))
   return average_word_lenght

  words = text.split()  #calculates the most repeated words
  word_frecuency = {}
  for word in words:
       word_frecuency[word] = word_frecuency.get (word, 0) + 1

  words_wanted = 10    #defines the number or repeated words we want to display
  most_repeated_words = sorted(word_frecuency.items(), key = lambda x: x[1], reverse = True)[:words_wanted]

  num_words = word_counter(text)
  num_sentences = sentence_counter (text)
  num_paragraphs =paragraph_counter(text)
  average_word_lenght = average_word_lenght(text)

  print("Text Analysis Tool\n")
  print("Number of words:",num_words)
  print("Number of sentences:",num_sentences)
  print("Number of paragraphs",num_paragraphs)
  print("Average word lenght",average_word_lenght)
  print("Most repeated words")
  for word, frecuency in most_repeated_words:
         print(f"'{word}' --> {frecuency} times in text")



```

## 3. Number Guessing Game
#### Create a program that generates a random number between 1 and 100, and then asks the user to guess the number. The program should provide hints to the user if the guessed number is greater or less than the target number, and keep asking for a new guess until the user guesses correctly.

```python
import random

print ("Number guessing game \n")
player_name = input ("Enter your name: ")
random_number = random.randint (1, 10
                                0) ##Chooses a random number between 1 and 1000
tries = 1
print(f"""\Hello {player_name}, lets play a guessing game. I will select a number between 1 and 1000 and you should try to guess it, you have only 10 tries. Good luck!""")

while tries <= 10: ## number of tries
  guessed_number = int(input('\nGuess the number: '))
  if guessed_number == random_number:
    if tries == 1:
      print (f'\n Congrats {player_name}, you guessed the number in the first try!') ## when you win at first try
    else:
      print (f'\n Congrats {player_name}, you guessed the number in {tries} tries!') ## when you win normally

      break

  elif tries == 10: # if you dont win
    pass

  elif guessed_number not in range (1, 101):  ## if the given number is not on the range
    print ('The given number is not in the range of 1 to 1000, please try with another number ')

  ##hints
  elif guessed_number > random_number:
    print('The given number is greater than the correct number, please try again')

  elif guessed_number < random_number:
        print('The given number is smaller than the correct one, please try again')
        tries += 1

  else:
    print (f"\n Ohhh no {player_name}, the correct number was {random_number}, am so sorry, good luck on your next try")
```
